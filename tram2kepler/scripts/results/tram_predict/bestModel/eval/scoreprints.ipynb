```python
import pandas as pd
```

```python
def makeTable(s: pd.DataFrame, name):
    mask = (s['P'] == 0) & (s['R'] == 0) & (s['F1'] == 0)
    s = s[~mask]

    s_to_sort = s.iloc[:-2]
    s_to_keep = s.iloc[-2:]

    s = s.sort_values(by=['P', 'R', 'F1'], ascending=False)

    s = pd.concat([s_to_sort, s_to_keep])

    s = s.round(3)
    s = s.fillna('-')

    print(s)
    s.to_latex(name + '.tex', index=False, escape=True, float_format="%.3f")
```

```python
makeTable(pd.read_csv('tune_s.csv'), 's')
makeTable(pd.read_csv('tune_s_aug.csv'), 'sa')
makeTable(pd.read_csv('tune_m.csv'), 'm')
makeTable(pd.read_csv('tune_m_aug.csv'), 'ma')
```
